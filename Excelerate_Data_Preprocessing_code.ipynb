```python
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])

def calculate_age(born):
    today = date.today()
    born = born.date() if hasattr(born, 'date') else born
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['Age'] = df['Date of Birth'].apply(calculate_age)

# 4. (Optional) Check the result
print(df[['Date of Birth', 'Age']].head())
df.drop(columns=['Date of Birth'],inplace=True)
```

```python
date_cols = [
    "Learner SignUp DateTime",
    "Opportunity End Date",
    "Entry created at",
    "Apply Date",
    "Opportunity Start Date"
]

# 1. Convert each to datetime (invalid parses → NaT)
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 2. Split out components
for col in date_cols:
    df[f"{col}_year"]   = df[col].dt.year
    df[f"{col}_month"]  = df[col].dt.month
    df[f"{col}_day"]    = df[col].dt.day
    df[f"{col}_hour"]   = df[col].dt.hour
    df[f"{col}_minute"] = df[col].dt.minute
    df[f"{col}_second"] = df[col].dt.second

# 3. Drop the original datetime columns
df.drop(columns=date_cols, inplace=True)
```

```python
suffixes = ['year', 'month', 'day', 'hour', 'minute', 'second']

for col in df.columns:
    if any(col.endswith(f"_{suf}") for suf in suffixes):
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
```
